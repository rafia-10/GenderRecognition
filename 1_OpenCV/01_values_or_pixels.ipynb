```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# create an array in the range (0,255)
arr = np.arange(0,256,1)
print(arr)
```

```python
arr = arr.reshape(16,16)
print(arr)
```

```python
arr_8bit = arr.astype(np.uint8)
```

```python
plt.imshow(arr_8bit, cmap='grey')
```

```python
# random array
randArr = np.random.randint(low= 0, high = 255, size= (16,16), dtype = np.uint8)
plt.imshow(randArr, cmap = 'grey')
```
